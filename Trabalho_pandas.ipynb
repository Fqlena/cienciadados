```python
def categorizar_valor(valor):
    if valor <= 90:
        return 'baixo'
    elif valor <= 120:
        return 'normal'
    else:
        return 'alto'

data_frame['Resultado Categorizado'] = data_frame['Resultado'].apply(categorizar_valor)
#salvar no novo csv
#data_frame.to_csv('/content/drive/MyDrive/Especialização/glicose_data_suja_trabalho_novo.csv', index=False)
#print(data_frame[['Dia Semana', 'Resultado']][:10])

#exclui linhas que tem uma coluna com algum valor nulo
data_frame = data_frame.dropna(how='any')




```
